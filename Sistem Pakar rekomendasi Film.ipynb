```python
# =========================================
# SISTEM PAKAR REKOMENDASI FILM
# TEXT-BASED USER INTERFACE (NUMERIC INPUT)
# =========================================

# Data film (title, duration, rating, genres, mood)
movies = [
    { "title": "Skyward Adventure", "duration": 145, "rating": 8.3, "genres": ["Petualangan","Keluarga"], "mood": ["senang","termotivasi"] },
    { "title": "Lighthearted Break", "duration": 85, "rating": 7.4, "genres": ["Komedi"], "mood": ["sedih","bosan","santai"] },
    { "title": "Midnight Mystery", "duration": 110, "rating": 8.0, "genres": ["Thriller","Misteri"], "mood": ["stres","termotivasi"] },
    { "title": "Cozy Romance", "duration": 95, "rating": 7.6, "genres": ["Romantis","Drama"], "mood": ["sedih","santai","senang"] },
    { "title": "Tiny Wonders", "duration": 28, "rating": 7.9, "genres": ["Dokumenter"], "mood": ["santai","termotivasi"] },
    { "title": "Epic Saga", "duration": 160, "rating": 8.6, "genres": ["Action","Petualangan"], "mood": ["senang","termotivasi"] },
    { "title": "The Secret Life of Walter Mitty", "duration": 140, "rating": 7.3, "genres": ["Petualangan"], "mood": ["senang","termotivasi"] },
    { "title": "Mr. Bean's Holiday", "duration": 90, "rating": 6.4, "genres": ["Komedi"], "mood": ["senang","santai"] },
    { "title": "Inside Out", "duration": 95, "rating": 8.1, "genres": ["Animasi"], "mood": ["sedih","bosan","santai"] },
    { "title": "The Shawshank Redemption", "duration": 142, "rating": 9.3, "genres": ["Drama"], "mood": ["sedih","termotivasi"] },
    { "title": "A Quiet Place", "duration": 90, "rating": 7.5, "genres": ["Horor"], "mood": ["stres","tegang"] },
    { "title": "Toy Story", "duration": 81, "rating": 8.3, "genres": ["Animasi"], "mood": ["senang","santai","remaja"] },
    { "title": "John Wick", "duration": 101, "rating": 7.4, "genres": ["Aksi"], "mood": ["termotivasi","senang"] },
    { "title": "The Intouchables", "duration": 112, "rating": 8.5, "genres": ["Drama"], "mood": ["sedih","santai"] },
    { "title": "Parasite", "duration": 132, "rating": 8.6, "genres": ["Drama"], "mood": ["stres","termotivasi"] },
    { "title": "Sing", "duration": 110, "rating": 7.1, "genres": ["Animasi"], "mood": ["senang","santai","remaja"] },
    { "title": "Interstellar", "duration": 169, "rating": 8.6, "genres": ["Sci-Fi","Drama"], "mood": ["termotivasi","stres"] },
    { "title": "La La Land", "duration": 128, "rating": 8.0, "genres": ["Romantis","Musikal"], "mood": ["senang","santai"] },
    { "title": "Zootopia", "duration": 108, "rating": 8.0, "genres": ["Animasi","Petualangan"], "mood": ["senang","bosan","remaja"] },
    { "title": "The Pursuit of Happyness", "duration": 117, "rating": 8.0, "genres": ["Drama"], "mood": ["sedih","termotivasi"] },
    { "title": "Coco", "duration": 105, "rating": 8.4, "genres": ["Animasi","Musikal"], "mood": ["senang","sedih"] },
    { "title": "Inception", "duration": 148, "rating": 8.8, "genres": ["Sci-Fi","Aksi"], "mood": ["termotivasi","stres"] },
    { "title": "The Martian", "duration": 144, "rating": 8.0, "genres": ["Sci-Fi"], "mood": ["termotivasi","senang"] },
    { "title": "Up", "duration": 96, "rating": 8.2, "genres": ["Animasi","Petualangan"], "mood": ["sedih","senang"] },
    { "title": "The Dark Knight", "duration": 152, "rating": 9.0, "genres": ["Aksi","Drama"], "mood": ["termotivasi","stres"] },
    { "title": "Ratatouille", "duration": 111, "rating": 8.1, "genres": ["Animasi","Komedi"], "mood": ["santai","senang"] },
    { "title": "Finding Nemo", "duration": 100, "rating": 8.1, "genres": ["Animasi","Petualangan"], "mood": ["senang","santai","remaja"] },
    { "title": "Joker", "duration": 122, "rating": 8.4, "genres": ["Drama","Thriller"], "mood": ["stres","termotivasi"] },
    { "title": "The Revenant", "duration": 156, "rating": 8.0, "genres": ["Drama","Petualangan"], "mood": ["stres","termotivasi"] },
    { "title": "Spirited Away", "duration": 125, "rating": 8.6, "genres": ["Animasi","Fantasi"], "mood": ["bosan","senang"] },
    { "title": "The Lion King", "duration": 88, "rating": 8.5, "genres": ["Animasi","Drama"], "mood": ["sedih","termotivasi"] },
    { "title": "Your Name", "duration": 106, "rating": 8.4, "genres": ["Animasi","Romantis"], "mood": ["senang","sedih"] },
    { "title": "Whiplash", "duration": 107, "rating": 8.5, "genres": ["Drama","Musikal"], "mood": ["termotivasi","stres"] },
    { "title": "Cinderella Man", "duration": 144, "rating": 8.0, "genres": ["Drama","Olahraga"], "mood": ["termotivasi","sedih"] },
    { "title": "Big Hero 6", "duration": 102, "rating": 7.8, "genres": ["Animasi","Aksi"], "mood": ["senang","santai","remaja"] },
    { "title": "The Grand Budapest Hotel", "duration": 99, "rating": 8.1, "genres": ["Komedi","Drama"], "mood": ["bosan","senang"] },
    { "title": "The Prestige", "duration": 130, "rating": 8.5, "genres": ["Drama","Mystery"], "mood": ["stres","termotivasi"] },
    { "title": "Life of Pi", "duration": 127, "rating": 7.9, "genres": ["Drama","Petualangan"], "mood": ["santai","termotivasi"] },
    { "title": "Spider-Man: Into the Spider-Verse", "duration": 117, "rating": 8.4, "genres": ["Animasi","Aksi"], "mood": ["senang","termotivasi"] },
    { "title": "The Social Network", "duration": 120, "rating": 7.7, "genres": ["Drama"], "mood": ["termotivasi","stres"] },
    { "title": "Moana", "duration": 107, "rating": 7.6, "genres": ["Animasi","Petualangan"], "mood": ["senang","santai"] },
    { "title": "Paddington", "duration": 95, "rating": 7.8, "genres": ["Keluarga","Komedi"], "mood": ["senang","santai"] },
    { "title": "The Maze Runner", "duration": 113, "rating": 6.8, "genres": ["Aksi","Sci-Fi"], "mood": ["tegang","termotivasi"] },
    { "title": "Home Alone", "duration": 103, "rating": 7.6, "genres": ["Komedi","Keluarga"], "mood": ["senang","santai"] },
    { "title": "Soul", "duration": 100, "rating": 8.0, "genres": ["Animasi","Drama"], "mood": ["tenang","senang"] },
    { "title": "1917", "duration": 119, "rating": 8.3, "genres": ["Drama","Perang"], "mood": ["tegang","stres"] },
    { "title": "Dune", "duration": 155, "rating": 8.0, "genres": ["Sci-Fi","Drama"], "mood": ["termotivasi","stres"] },
    { "title": "The Truman Show", "duration": 103, "rating": 8.1, "genres": ["Drama","Komedi"], "mood": ["santai","termotivasi"] },
    { "title": "Shutter Island", "duration": 138, "rating": 8.2, "genres": ["Thriller","Misteri"], "mood": ["tegang","stres"] },
    { "title": "WALL-E", "duration": 98, "rating": 8.4, "genres": ["Animasi","Sci-Fi"], "mood": ["senang","santai"] },
    { "title": "Weathering With You", "duration": 112, "rating": 7.5, "genres": ["Animasi","Romantis","Fantasi"], "mood": ["senang","sedih","tenang"] },
    { "title": "Forrest Gump", "duration": 142, "rating": 8.8, "genres": ["Drama","Romantis"], "mood": ["sedih","termotivasi"] },
    { "title": "Gladiator", "duration": 155, "rating": 8.5, "genres": ["Aksi","Drama"], "mood": ["termotivasi","tegang"] },
    { "title": "Avengers: Endgame", "duration": 181, "rating": 8.4, "genres": ["Aksi","Sci-Fi"], "mood": ["senang","termotivasi"] },
    { "title": "Frozen", "duration": 102, "rating": 7.4, "genres": ["Animasi","Musikal"], "mood": ["senang","santai","remaja"] },
    { "title": "How to Train Your Dragon", "duration": 98, "rating": 8.1, "genres": ["Animasi","Petualangan"], "mood": ["senang","termotivasi"] },
    { "title": "Mad Max: Fury Road", "duration": 120, "rating": 8.1, "genres": ["Aksi","Petualangan"], "mood": ["tegang","termotivasi"] },
    { "title": "The Notebook", "duration": 123, "rating": 7.8, "genres": ["Romantis","Drama"], "mood": ["sedih","santai"] },
    { "title": "Her", "duration": 126, "rating": 8.0, "genres": ["Drama","Sci-Fi"], "mood": ["tenang","sedih"] },
    { "title": "Gravity", "duration": 91, "rating": 7.7, "genres": ["Sci-Fi","Thriller"], "mood": ["tegang","stres"] },
    { "title": "Deadpool", "duration": 108, "rating": 8.0, "genres": ["Aksi","Komedi"], "mood": ["senang","bosan"] },
    { "title": "Logan", "duration": 137, "rating": 8.1, "genres": ["Aksi","Drama"], "mood": ["sedih","termotivasi"] },
    { "title": "The Imitation Game", "duration": 114, "rating": 8.0, "genres": ["Drama"], "mood": ["termotivasi","sedih"] },
    { "title": "Black Panther", "duration": 134, "rating": 7.3, "genres": ["Aksi","Sci-Fi"], "mood": ["senang","termotivasi"] },
    { "title": "Doctor Strange", "duration": 115, "rating": 7.5, "genres": ["Aksi","Fantasi"], "mood": ["senang","termotivasi"] },
    { "title": "The Hunger Games", "duration": 142, "rating": 7.2, "genres": ["Sci-Fi","Drama"], "mood": ["tegang","termotivasi"] },
    { "title": "Ready Player One", "duration": 140, "rating": 7.4, "genres": ["Sci-Fi","Petualangan"], "mood": ["senang","bosan"] },
    { "title": "The Fault in Our Stars", "duration": 126, "rating": 7.7, "genres": ["Romantis","Drama"], "mood": ["sedih","tenang"] },
    { "title": "The Conjuring", "duration": 112, "rating": 7.5, "genres": ["Horor","Thriller"], "mood": ["tegang","stres"] },
    { "title": "A Beautiful Mind", "duration": 135, "rating": 8.2, "genres": ["Drama"], "mood": ["sedih","termotivasi"] },
    { "title": "Slumdog Millionaire", "duration": 120, "rating": 8.0, "genres": ["Drama"], "mood": ["termotivasi","senang"] },
    { "title": "The Wolf of Wall Street", "duration": 180, "rating": 8.2, "genres": ["Drama","Komedi"], "mood": ["senang","termotivasi"] },
    { "title": "Arrival", "duration": 116, "rating": 7.9, "genres": ["Sci-Fi","Drama"], "mood": ["tenang","stres"] },
    { "title": "The Help", "duration": 146, "rating": 8.1, "genres": ["Drama"], "mood": ["sedih","termotivasi"] },
    { "title": "Nightcrawler", "duration": 118, "rating": 7.8, "genres": ["Thriller","Drama"], "mood": ["stres","tegang"] },
    { "title": "The King's Speech", "duration": 118, "rating": 8.0, "genres": ["Drama"], "mood": ["termotivasi","tenang"] },
    { "title": "Edge of Tomorrow", "duration": 113, "rating": 7.9, "genres": ["Aksi","Sci-Fi"], "mood": ["tegang","senang"] },
    { "title": "The Lego Movie", "duration": 100, "rating": 7.7, "genres": ["Animasi","Komedi"], "mood": ["senang","bosan"] },
    { "title": "Cast Away", "duration": 143, "rating": 7.8, "genres": ["Drama","Petualangan"], "mood": ["sedih","termotivasi"] },
    { "title": "Shrek", "duration": 90, "rating": 7.9, "genres": ["Animasi","Komedi"], "mood": ["senang","santai"] },
    { "title": "The Iron Giant", "duration": 86, "rating": 8.0, "genres": ["Animasi","Petualangan"], "mood": ["sedih","senang"] }
]



# =========================================
# OPSI INPUT (SEMUA ANGKA)
# =========================================

mood_options = {
    1: "senang",
    2: "sedih",
    3: "bosan",
    4: "stres",
    5: "santai",
    6: "termotivasi"
}

time_options = {
    1: "<60",
    2: "60-120",
    3: ">120"
}

genre_options = {
    1: "Animasi",
    2: "Aksi",
    3: "Drama",
    4: "Komedi",
    5: "Petualangan",
    6: "Sci-Fi",
    7: "Romantis",
    8: "Thriller",
    9: "Dokumenter",
    10: "Keluarga"
}

# =========================================
# KLASIFIKASI WAKTU
# =========================================

def classify_time(minutes):
    if minutes < 60:
        return "<60"
    elif minutes <= 120:
        return "60-120"
    else:
        return ">120"

# =========================================
# RULE BASE (10 HIPOTESIS)
# =========================================

rules = [
    {"if": {"mood": "sedih", "time": "<60", "genre": "Animasi"},
     "then": "Rekomendasi Film Animasi Pendek"},

    {"if": {"mood": "bosan", "time": "60-120", "genre": "Komedi"},
     "then": "Rekomendasi Film Komedi Sedang"},

    {"if": {"mood": "sedih", "time": "60-120", "genre": "Drama"},
     "then": "Rekomendasi Film Drama Sedang"},

    {"if": {"mood": "termotivasi", "time": ">120", "genre": "Aksi"},
     "then": "Rekomendasi Film Aksi Panjang"},

    {"if": {"mood": "senang", "time": ">120", "genre": "Petualangan"},
     "then": "Rekomendasi Film Petualangan Panjang"},

    {"if": {"mood": "stres", "time": "60-120", "genre": "Thriller"},
     "then": "Rekomendasi Film Thriller Sedang"},

    {"if": {"mood": "senang", "time": "60-120", "genre": "Romantis"},
     "then": "Rekomendasi Film Romantis Sedang"},

    {"if": {"mood": "santai", "time": "<60", "genre": "Dokumenter"},
     "then": "Rekomendasi Film Dokumenter Pendek"},

    {"if": {"mood": "termotivasi", "time": ">120", "genre": "Sci-Fi"},
     "then": "Rekomendasi Film Sci-Fi Panjang"},

    {"if": {"mood": "senang", "time": "60-120", "genre": "Keluarga"},
     "then": "Rekomendasi Film Keluarga Sedang"}
]

# =========================================
# INFERENCE ENGINE (FORWARD CHAINING)
# =========================================

def infer(mood, time_cat, genre):
    for rule in rules:
        c = rule["if"]
        if c["mood"] == mood and c["time"] == time_cat and c["genre"] == genre:
            return rule["then"]
    return None

# =========================================
# SISTEM UTAMA
# =========================================

while True:
    print("\n=== SISTEM PAKAR REKOMENDASI FILM ===")

    print("\nPilih Mood:")
    for k, v in mood_options.items():
        print(f"{k}. {v}")
    mood = mood_options.get(int(input("Masukkan nomor mood: ")))
    if not mood:
        print("Input tidak valid.")
        continue

    print("\nPilih Durasi Film:")
    print("1. < 60 menit")
    print("2. 60 - 120 menit")
    print("3. > 120 menit")
    time_cat = time_options.get(int(input("Masukkan nomor durasi: ")))
    if not time_cat:
        print("Input tidak valid.")
        continue

    print("\nPilih Genre:")
    for k, v in genre_options.items():
        print(f"{k}. {v}")
    genre = genre_options.get(int(input("Masukkan nomor genre: ")))
    if not genre:
        print("Input tidak valid.")
        continue

    conclusion = infer(mood, time_cat, genre)

    print("\n=== HASIL SISTEM PAKAR ===")
    if conclusion:
        print("KESIMPULAN:", conclusion)
    else:
        print("KESIMPULAN: Tidak ditemukan aturan yang sesuai")

    print("\nFilm yang direkomendasikan:")
    found = False
    for m in movies:
        if genre in m["genres"] and mood in m["mood"] and classify_time(m["duration"]) == time_cat:
            print(f"- {m['title']} ({m['duration']} menit)")
            found = True

    if not found:
        print("Tidak ada film yang cocok.")

    print("\n1. Cari lagi")
    print("2. Keluar")
    if int(input("Pilih: ")) != 1:
        print("Program selesai. Terima kasih.")
        break
```
